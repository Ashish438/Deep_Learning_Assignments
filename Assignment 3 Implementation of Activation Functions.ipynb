```python
"""

Activation Functions are implemented:
1) Linear
2) ReLU
3) Leaky ReLU
4) Parametric ReLU
5) Sigmoid Function
6) TanH
7) Softmax

"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Linear Activation Function"""
x = np.arange(-10,10,0.5)
plt.plot(x)
plt.grid()
```

```python

"""ReLU (Rectified Linear Unit) Function"""
def relu(x):
    a = []
    for item in x:
        if item > 0:
            a.append(item)
        else:
            a.append(0)
    return a

y = relu(x)
plt.plot(x,y)
plt.grid()
```

```python
"""Leaky ReLU Activation Function"""
def leaky_relu(x):
    b = []
    for item in x:
        if item > 0:
            b.append(item)
        else:
            item = item * 0.01
            b.append(item)
    return b

y = leaky_relu(x)
plt.plot(x,y)
plt.grid()
```

```python
"""Parametric ReLU Activation Function"""
def parameterized_relu(x):
    a = []
    for item in x:
        if item > 0:
            a.append(item)
        else:
            item = item * 2
            a.append(item)
    return a

y = parameterized_relu(x)
plt.plot(x,y)
plt.grid()
```

```python
"""Sigmoid / Logistic Activation Function"""
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)
plt.plot(x,y)
plt.grid()
```

```python

"""TanH / Hyperbolic Tangent Activation Function"""

def tanh(x):
    a = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
    return a

y = tanh(x)
plt.plot(x,y)
plt.grid()
```

```python
"""Softmax Activation Function"""
def softmax_function(x):
    z = np.exp(x)
    z_ = z/z.sum()
    return z_

y = softmax_function(x)
plt.plot(x,y)
plt.grid()
```
